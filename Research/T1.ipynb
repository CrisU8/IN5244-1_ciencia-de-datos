# Pregunta 4

Test de medias

```python
from scipy import stats
```

```python
def perform_t_tests(var_list, group1, group2, alpha=0.1):
    for var in var_list:
        t_stat, p_value = stats.ttest_ind(group1[var].dropna(), group2[var].dropna(), equal_var=False)
        print(f"Variable: {var}")
        print(f"Estadístico t: {t_stat}")
        print(f"Valor p: {p_value}")
        
        if p_value < alpha:
            print("Se rechaza la hipótesis nula: las medias de los dos grupos son significativamente diferentes.")
        else:
            print("No se rechaza la hipótesis nula: no hay evidencia suficiente para afirmar que las medias de los dos grupos son significativamente diferentes.")
        print() 
```

```python
# Realizar los test t para todas las variables
variables = ["Age","GPAin2021","MunicipalSchool","AttendanceIn2021","Female"]
perform_t_tests(variables, migrantes, no_migrantes)
```

Considerando $$y_i = \beta_o + \beta_{1}T_i  + \epsilon_i$$

a)$E[y_1i | T_i = 1]$: Esta expresión es la esperanza del resultado de revinculación para y bajo tratamiento dado que lo recibio, esto es observable.

b)$E[y_0i | T_i = 1]$: Esta expresión es la esperanza del resultado de revinculación para y bajo control dado que recibe tratamiento, esto es no observable.

c)$E[y_1i | T_i = 0]$: Esta expresión es la esperanza del resultado de revinculación para y bajo tratamiento dado que **no** recibe tratamiento. Esto es no observable.

d)$E[y_0i | T_i = 0]$: Esta expresión es la espranza del resultado de revinculación para y bajo control dado que no recibe tratamiento. Esto es observable.


# Pregunta 9

\begin{equation}
    \begin{split}
        \beta_{1} &= E[y_i \mid D=1] - E[y_i \mid D=0] \\
        \beta_{1} &= E[y_{1i} \mid D=1] - E[y_{0i} \mid D=0] \\
        \beta_{1} &= E[y_{1i} \mid D=1] - E[y_{0i} \mid D=0] 
                   + E[y_{0i} \mid D=1] - E[y_{0i} \mid D=1] \\
        \beta_{1} &= \underbrace{E[y_{1i} - y_{0i} \mid D=1]}_{\text ATET} + \underbrace{E[y_{0i} \mid D=1] - E[y_{0i} \mid D=0]}_{\text{Sesgo de selección}} 
    \end{split}
\end{equation}

ATET: Efecto promedio del tratamiento en los tratados.

Tomando entonces `ListenInDropoutReport` como la variable de tratamiento, y `EnrolledByAug212022` como la variable $y_i$.

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
from statsmodels.formula.api import ols
from scipy import stats
```

```python
formula = 'EnrolledByAug312022 ~ ListedInDropoutReport'
model = smf.ols(formula, data=df).fit()

# Obtener el resumen del modelo
summary = model.summary()

# Mostrar el resumen
print(summary)

# Convertir el resumen en una tabla
results_df = pd.DataFrame({
    'Coeficientes': model.params,
    'Error Estándar': model.bse,
    'Estadístico t': model.tvalues,
    'Valor p': model.pvalues
})

# Mostrar la tabla de resultados
print("\nTabla de Resultados:")
print(results_df)
```

```python
# Ajusta el modelo de regresión
model = ols('EnrolledByAug312022 ~ ListedInDropoutReport', data=df).fit()

# Ajustar errores estándar clusterizados a nivel de 'IDSchool'
clustered_se = model.get_robustcov_results(cov_type='cluster', groups=df['IDschool'])

# Resumen del modelo con errores estándar clusterizados
print(clustered_se.summary())
```

```python
# Dividir los datos en grupos de tratamiento y control
treatment_group = df[df['ListedInDropoutReport'] == 1]['EnrolledByAug312022']
control_group = df[df['ListedInDropoutReport'] == 0]['EnrolledByAug312022']

# Realizar el test t
t_stat, p_value = stats.ttest_ind(treatment_group, control_group, equal_var=True)

print("\nResultado del Test t de Medias:")
print(f'Estadístico t: {t_stat}')
print(f'Valor p: {p_value}')
```
